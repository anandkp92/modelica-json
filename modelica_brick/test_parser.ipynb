```python
from modelica_brick_parser import *
import datetime
```

```python
parser = Modelica_Brick_Parser(modelica_buildings_library_path='Buildings.Examples.VAVReheat', modelica_json_filename='Guideline36', json_folder='raw-json')
```

```python
brick_relationships = parser.get_brick_relationships()
```

```python
records = []
for relationship in brick_relationships:
    brick_record = {
        "timestamp": datetime.datetime.utcnow(),
        "triples": [
            relationship['obj1'], relationship['relationship'], relationship['obj2']
        ]
    }
    records.append(brick_record)
```
